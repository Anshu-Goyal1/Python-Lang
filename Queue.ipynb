```python
Q[]
rear=0
size=int (input("please enter the size of the Queue"))
```

```python
print("1.EnQueue\n2.Dequeue\n3.Display\n4.Exit")
choice=int (input("Please enter your choice "))
if choice==1:
  if rear==size:
    print("Over Flow")
  else:
    data=int (input("Please enter your data for EnQueue " " "))
  Q.append(data)
  print(data," Enqueued " "")
elif choice==2:
  if Q==[]:
    print("Under flow")
  else:
    data=Q.pop(0)
    print(data,"Dequeued ")
elif choice==3:
  if Q==[]:
    print("Queue is Empty ")
  for T in Q:
    print(T,"<--",end=" ")
 
  print("*"*20,"\n\n\n")
elif choice==4:
  pass
else:
    print("Invalid option ")
```
