```python
#reading a text file using a clause for counting no. of words
with open('National Anthem','r') as f:
  data=f.readlines()
no_of_words=0
for lines in data:
 words=lines.split()
 for word in words:
    no_of_words+=1
    print(word)
    print(no_of_words,"words")
```
